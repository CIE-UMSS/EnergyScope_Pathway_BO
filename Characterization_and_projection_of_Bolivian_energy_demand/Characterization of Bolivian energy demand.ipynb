```python
# Import the Google Drive Excel file
google_drive_url = 'https://docs.google.com/spreadsheets/d/1vzgPGMI4y3XHQMNRRJsmn5R7IPVXhnqv/edit?usp=sharing&ouid=107031919989880356069&rtpof=true&sd=true'
# Find the index of '/edit'
edit_index = google_drive_url.find('/edit')
# Remove all text after '/edit' (including '/edit')
google_drive_url = google_drive_url[:edit_index]
# Append '/export' to the cleaned URL
google_drive_url = google_drive_url + '/export'
# Read the Excel file into a Pandas DataFrame
df = pd.read_excel(google_drive_url, header=4, engine='openpyxl')
```

```python
# Exclude rows 0, 29, 30, 31
rows_to_exclude = [0, 29, 30, 31]
rows_to_exclude = [idx for idx in rows_to_exclude if idx in df.index]
df = df.drop(rows_to_exclude)
```

```python
# Modify the first column name
df.columns.values[0] = ''
```

```python
# Convert TJ (10^12 J) to GWh
df.iloc[:, 1:] = df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce') * 0.2777777777777777777777777777777777777777777777778
```

```python
# Replace NaN values with blank space
df = df.fillna('')
```

```python
# Reset the index
df = df.reset_index(drop=True)
```

```python
# Use the first column as row names
df.set_index(df.iloc[:, 0], inplace=True)
```

```python
# Delete the first column
df.drop(df.columns[[0]], axis=1, inplace=True)
```

```python
# Display the DataFrame
df
```

```python
# Find the row with a specific name
selected_row = df[df.index.str.contains("TRANSPORT")]

# Find the headers for the specified columns
selected_columns = ["NATURAL GAS", "ELECTRICITY", "GASOLINE WITHOUT ETHANOL"]
headers_containing_cols = {}
for col in selected_columns:
    headers_containing_cols[col] = [header for header in selected_row.columns if col in header]

# Create a DataFrame for the specified row and columns
data = {}
for col, headers_containing_col in headers_containing_cols.items():
    if headers_containing_col:
        data[col] = selected_row[headers_containing_col].iloc[0].values

result_df = pd.DataFrame(data, index=[selected_row.index[0]])

# Print the new DataFrame
result_df
```

<font size="3"><b>3.1.1. Cochabamba department</b></font> <a class="anchor" id="section_3_1_1"></a>

```python
# Number of households with electricity and water distribution in Cochabamba
num_households_WD_CB=557331
# Share of households in Cochabamba that possess appliances for water heating
share_household_WH_CB=0.6381
# Energy demand related to water heating per household for Cochabamba (GWh/household)
household_WH_EUD_CB=0.00052369
# Efficiency of solar thermal panels
Eff_RES_SOLAR_panel=1
# Energy demand related to solar thermal systems for Cochabamba (GWh)
RES_SOLAR_WH_FEC_CB=5.2920135
# End-use demand related to solar thermal systems for Cochabamba (GWh)
RES_SOLAR_WH_EUD_CB=RES_SOLAR_WH_FEC_CB*Eff_RES_SOLAR_panel
# Share of electric showers related to water heating
share_electric_shower=0.975
# Efficiency of electric showers
Eff_EL_shower=1
# Electriciy demand for water heating in Cochabamba (GWh)
EL_WH_FEC_CB=((num_households_WD_CB*share_household_WH_CB*household_WH_EUD_CB)-RES_SOLAR_WH_EUD_CB)*share_electric_shower
# End-use demand related to electric showers for Cochabamba (GWh)
EL_WH_EUD_CB=EL_WH_FEC_CB*Eff_EL_shower
# Efficiency of fossil gas boiler
Eff_FG_boiler=1/1.121681905
# Fossil gas demand for water heating in Cochabamba (GWh)
FG_WH_FEC_CB=(((num_households_WD_CB*share_household_WH_CB*household_WH_EUD_CB)-RES_SOLAR_WH_EUD_CB)*(1-share_electric_shower))/Eff_FG_boiler
# End-use demand related to fossil gas boiler for Cochabamba (GWh)
FG_WH_EUD_CB=FG_WH_FEC_CB*Eff_FG_boiler
```

```python
# Make a copy of the DataFrame
copied_df = End_Use_df.copy()

# Convert non-numeric values in all numeric columns to NaN
numeric_cols = ['EL', 'FG', 'LPG', 'WD', 'RES_SOLAR', 'DS', 'GS', 'LFO']
copied_df[numeric_cols] = copied_df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Convert empty values to NaN in the first 5 columns
copied_df.iloc[:, :5] = copied_df.iloc[:, :5].applymap(lambda x: x if x != '' else pd.NA)

# Create a new column 'Division'
copied_df['Division'] = ''

# Iterate through each row
for index, row in copied_df.iterrows():
    # Iterate through the first 5 columns in reverse order
    for i in range(4, -1, -1):
        if pd.notna(row[i]):
          if row[0] == 'Transport':
            copied_df.at[index, 'Division'] = row[i]
          elif row[0] == 'Residential' and row[1] == 'Other Appliances':
            copied_df.at[index, 'Division'] = row[i]
          elif row[0] == 'Public Lighting':
            copied_df.at[index, 'Division'] = row[i]
          elif row[0] == 'Commercial' and row[1] == 'Other Appliances':
            copied_df.at[index, 'Division'] = row[i]
          elif row[0] == 'Industrial' and row[1] == 'Cement':
            copied_df.at[index, 'Division'] = row[i-2]
          elif row[0] == 'Industrial' and row[2] == 'MSMEs':
            copied_df.at[index, 'Division'] = row[i-2]
          else:
            copied_df.at[index, 'Division'] = row[i-1]
          break
```

```python
# Filter rows where 'Sector' column contains 'Transport'
filtered_df = copied_df[copied_df['Sector'].str.contains('Transport')]

# Group by 'Division' and sum the values for each header
grouped_df = filtered_df.groupby('Division')[['EL', 'FG', 'LPG', 'WD', 'RES_SOLAR', 'DS', 'GS', 'LFO']].sum()

# Exclude headers with no values
grouped_df = grouped_df.loc[:, (grouped_df != 0).any(axis=0)]

# Sort the DataFrame by the sum of values in each row in descending order
grouped_df = grouped_df.loc[grouped_df.sum(axis=1).sort_values(ascending=False).index]

# Reset the index to make 'Division' a regular column
grouped_df = grouped_df.reset_index()

# Melt the DataFrame to create a long format for plotly express
melted_df = pd.melt(grouped_df, id_vars='Division', var_name='Header', value_name='Values')

# Create a stacked bar plot using plotly express
fig = px.bar(melted_df, x='Header', y='Values', color='Division',
             #title='Stacked Bar Plot by Division for Residential Sector',
             labels={'Values': 'FEC (GWh)', 'Header': ''})

# Update layout for better readability
fig.update_layout(
    height=700,
    width=1375
)

# Update legend title
fig.update_layout(
    showlegend=True,
    legend_title_text='Technology',
    legend_title_font=dict(size=12.5)
)

# Show the plot
fig.show(renderer="iframe")
```

```python
# Option2: Import the Google Drive Excel file
google_drive_url2 = 'https://docs.google.com/spreadsheets/d/1HvZ6PBKkdaFy-yR9gaXc1wu4jfgGn_Ce/edit?usp=drive_link&ouid=115521570694102221364&rtpof=true&sd=true'
# Find the index of '/edit'
edit_index2 = google_drive_url2.find('/edit')
# Remove all text after '/edit' (including '/edit')
google_drive_url2 = google_drive_url2[:edit_index]
# Append '/export' to the cleaned URL
google_drive_url2 = google_drive_url2 + '/export'
# Read the Excel file into a Pandas DataFrame
df2 = pd.read_excel(google_drive_url2, header=0, engine='openpyxl')
```

```python
# Renaming technology column to Final-demand
df2.rename(columns={'Unnamed: 0': 'Final-demand'}, inplace=True)

# Grouping up end use categories that can be grouped up
group_list = {'Conservacion de alimentos', 'Aire acondicionado', 'Ventilacion', 'Bombeo de agua', 'Artefactos electricos'}
column_to_check = 'Final-demand'

selected_rows = pd.DataFrame(columns=df2.columns)

for tech in df2[column_to_check].unique():
    if tech in group_list:
        # Filter rows based on 'Final-demand' value
        tech_rows = df2[df2[column_to_check] == tech]

        # Sum the filtered rows
        aggregated_row = tech_rows

        # Append the aggregated row to the result DataFrame
        selected_rows = pd.concat([selected_rows, aggregated_row], ignore_index=True)

new_row = selected_rows.sum(numeric_only=True)
new_row['Final-demand'] = 'Electrical_appliances'

df2 = pd.concat([df2, new_row.to_frame().T], ignore_index=True)
```

```python
# Make a copy of the DataFrame
copied_df = End_Use_df.copy()

# Convert non-numeric values in all numeric columns to NaN
numeric_cols = ['EL', 'FG', 'LPG', 'WD', 'RES_SOLAR', 'DS', 'GS', 'LFO']
copied_df[numeric_cols] = copied_df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Convert empty values to NaN in the first 5 columns
copied_df.iloc[:, :5] = copied_df.iloc[:, :5].applymap(lambda x: x if x != '' else pd.NA)

# Create a new column 'Division'
copied_df['Division'] = ''

# Iterate through each row
for index, row in copied_df.iterrows():
    # Iterate through the first 5 columns in reverse order
    for i in range(4, -1, -1):
        if pd.notna(row[i]):
          if row[0] == 'Transport':
            copied_df.at[index, 'Division'] = row[i]
          elif row[0] == 'Residential' and row[1] == 'Other Appliances':
            copied_df.at[index, 'Division'] = row[i]
          elif row[0] == 'Public Lighting':
            copied_df.at[index, 'Division'] = row[i]
          elif row[0] == 'Commercial' and row[1] == 'Other Appliances':
            copied_df.at[index, 'Division'] = row[i]
          elif row[0] == 'Industrial' and row[1] == 'Cement':
            copied_df.at[index, 'Division'] = row[i-2]
          elif row[0] == 'Industrial' and row[2] == 'MSMEs':
            copied_df.at[index, 'Division'] = row[i-2]
          else:
            copied_df.at[index, 'Division'] = row[i-1]
          break
```

<font size="3"><b>7.3.1. Bar chart - Final energy consumption by type of usage</b></font> <a class="anchor" id="section_7_3_1"></a>

```python
# Make a copy of the DataFrame
copied_df = End_Use_df.copy()

# Convert non-numeric values in all numeric columns to NaN
numeric_cols = ['EL', 'FG', 'LPG', 'WD', 'RES_SOLAR', 'DS', 'GS', 'LFO']
copied_df[numeric_cols] = copied_df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Convert empty values to NaN in the first 5 columns
copied_df.iloc[:, :5] = copied_df.iloc[:, :5].applymap(lambda x: x if x != '' else pd.NA)

# Create a new column 'Division'
copied_df['Division'] = ''

# Iterate through each row
for index, row in copied_df.iterrows():
    # Iterate through the first 5 columns in reverse order
    for i in range(4, -1, -1):
        if pd.notna(row[i]):
          if row[0] == 'Transport':
            copied_df.at[index, 'Division'] = row[i]
          elif row[0] == 'Residential' and row[1] == 'Other Appliances':
            copied_df.at[index, 'Division'] = row[i]
          elif row[0] == 'Public Lighting':
            copied_df.at[index, 'Division'] = row[i]
          elif row[0] == 'Commercial' and row[1] == 'Other Appliances':
            copied_df.at[index, 'Division'] = row[i]
          elif row[0] == 'Industrial' and row[1] == 'Cement':
            copied_df.at[index, 'Division'] = row[i-2]
          elif row[0] == 'Industrial' and row[2] == 'MSMEs':
            copied_df.at[index, 'Division'] = row[i-2]
          else:
            copied_df.at[index, 'Division'] = row[i-1]
          break
```

```python
# Filter rows where 'Sector' column contains 'Commercial'
filtered_df = copied_df[copied_df['Sector'].str.contains('Industrial')]

# Group by 'Division' and sum the values for each header
grouped_df = filtered_df.groupby('Division')[['EL', 'FG', 'LPG', 'WD', 'RES_SOLAR', 'DS', 'GS', 'LFO']].sum()

# Exclude headers with no values
grouped_df = grouped_df.loc[:, (grouped_df != 0).any(axis=0)]

# Sort the DataFrame by the sum of values in each row in descending order
grouped_df = grouped_df.loc[grouped_df.sum(axis=1).sort_values(ascending=False).index]

# Reset the index to make 'Division' a regular column
grouped_df = grouped_df.reset_index()

# Melt the DataFrame to create a long format for plotly express
melted_df = pd.melt(grouped_df, id_vars='Division', var_name='Header', value_name='Values')

# Create a stacked bar plot using plotly express
fig = px.bar(melted_df, x='Header', y='Values', color='Division',
             #title='Stacked Bar Plot by Division for Residential Sector',
             labels={'Values': 'FEC (GWh)', 'Header': ''})

# Update layout for better readability
fig.update_layout(
    height=700,
    width=1375
)

# Update legend title
fig.update_layout(
    showlegend=True,
    legend_title_text='Type of usage',
    legend_title_font=dict(size=12.5)
)

# Show the plot
fig.show(renderer="iframe")
```

```python
# Filter rows where 'Sector' column contains 'Commercial'
filtered_df = copied_df[copied_df['Sector'].str.contains('Non-energy')]

# Group by 'Division' and sum the values for each header
grouped_df = filtered_df.groupby('Division')[['EL', 'FG', 'LPG', 'WD', 'RES_SOLAR', 'DS', 'GS', 'LFO']].sum()

# Exclude headers with no values
grouped_df = grouped_df.loc[:, (grouped_df != 0).any(axis=0)]

# Sort the DataFrame by the sum of values in each row in descending order
grouped_df = grouped_df.loc[grouped_df.sum(axis=1).sort_values(ascending=False).index]

# Reset the index to make 'Division' a regular column
grouped_df = grouped_df.reset_index()

# Melt the DataFrame to create a long format for plotly express
melted_df = pd.melt(grouped_df, id_vars='Division', var_name='Header', value_name='Values')

# Create a stacked bar plot using plotly express
fig = px.bar(melted_df, x='Header', y='Values', color='Division',
             #title='Stacked Bar Plot by Division for Residential Sector',
             labels={'Values': 'FEC (GWh)', 'Header': ''})

# Update layout for better readability
fig.update_layout(
    height=700,
    width=1375
)

# Update legend title
fig.update_layout(
    showlegend=True,
    legend_title_text='Type of usage',
    legend_title_font=dict(size=12.5)
)

# Show the plot
fig.show(renderer="iframe")
```
